```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory'

html = requests.get(url)
html.status_code
```

```python
soup = BeautifulSoup(html.text, 'html5lib')
```

```python
soup.title
```

```python
html_tables = soup.find_all('div', class_='scroll-container')
```

```python
first_table = html_tables[0]
```

```python
print(first_table)
```

```python
column = first_table.find_all('th')

column_name = []
for header in column[:5]:
    header_text = header.text.strip()
    if header_text:  # Check if the header text is not empty
        column_name.append(header_text)
```

```python
print(column_name)
```

```python
covid_dict = dict.fromkeys(column_name)

covid_dict['Country'] = []
covid_dict['Deaths / million'] = []
covid_dict['Deaths'] = []
covid_dict['Cases'] = []
```

```python
covid_dict
```

```python
for name in column[5:]:
    name_text = name.text.strip()
    if name_text:
        covid_dict['Country'].append(name_text)
```

```python
covid_dict['Country']
```

```python
deth = covid_dict['Country']
print(len(deth))
```

```python
column_data = first_table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data if data.text.strip()]
    if len(individual_row_data) == 3:
        # Convert the number to int and handle the '' (em dash) as null
        deaths_per_million_str = individual_row_data[0].replace(',', '')
        deaths_per_million = int(deaths_per_million_str) if deaths_per_million_str != '' else None
        covid_dict['Deaths / million'].append(deaths_per_million)
    else:
        print("Skipping incomplete row:", individual_row_data)
```

```python
deth = covid_dict['Deaths / million']
print(len(deth))
```
